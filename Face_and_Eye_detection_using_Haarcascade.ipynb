# **Face Detection using Haarcascade**

```python
# Image as i/p for detection
from google.colab.patches import cv2_imshow
import cv2
sample=cv2.imread(r'/content/object detection assignment/Modi.jpg')
cv2_imshow(sample)
```

```python
# We point OpenCV's CascadeClassifier function to where our classifier (XML file format) is stored
face_cascade=cv2.CascadeClassifier(r'/content/object detection assignment/haarcascade_frontalface_default.xml')
```

```python
# Convert image to grayscale as i/p for haarcascade models is gray
gray=cv2.cvtColor(sample,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
# Function for face detection

def face_detection(frame):
  # Our classifier returns the ROI of the detected face as a tuple
  # It stores the top left coordinate and the bottom right coordiantes
  faces=face_cascade.detectMultiScale(gray,1.1,4)
  # When no faces detected, face_classifier returns and empty tuple
  if faces is ():
     print("No faces found")
  # We iterate through our faces array and draw a rectangle
  # over each face in faces
  for (x,y,w,h) in faces:
    cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),3)
    font=cv2.FONT_HERSHEY_DUPLEX
    cv2.putText(frame,'person',(x+6,y-6),font,0.5,(255,0,0),1)
  return frame
```

```python
# Give i/p image to function face_detection and find the o/p
from google.colab.patches import cv2_imshow
output=face_detection(sample)
cv2_imshow(output)
```

```python
# Check for another image with same model
sample2=cv2.imread(r'/content/object detection assignment/sample.jpg')
output=face_detection(sample2)
cv2_imshow(output)
```

# **Face and Eye detection using Haarcascade**

```python
eye_classifier = cv2.CascadeClassifier('/content/object detection assignment/haarcascade_eye.xml')
 
# Function for face and eye detection

def face_eye_detection(frame):
  # Our classifier returns the ROI of the detected face as a tuple
  # It stores the top left coordinate and the bottom right coordiantes
  faces=face_cascade.detectMultiScale(gray,1.1,4)
  # When no faces detected, face_classifier returns and empty tuple
  if faces is ():
     print("No faces found")
  # We iterate through our faces array and draw a rectangle
  # over each face in faces
  for (x,y,w,h) in faces:
    cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),3)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = sample[y:y+h, x:x+w]
    eyes = eye_classifier.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(255,255,0),2)
        # cv2_imshow(img)
    font=cv2.FONT_HERSHEY_DUPLEX
    cv2.putText(frame,'person',(x+6,y-6),font,0.5,(255,0,0),1)
  return frame
```

```python
# Give i/p image to function face_detection and find the o/p
from google.colab.patches import cv2_imshow
output=face_eye_detection(sample)
cv2_imshow(output)
```
